```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Load Data

```python
path = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(path)
```

```python
titanic.head()
```

Check dataset info

```python
titanic.info()
```

### Visualizations

Distribution of Survived

```python
sns.set_style('whitegrid')
```

```python
survived = sns.countplot(data= titanic, x= 'Survived'); # 0 = died, 1 = survived
plt.title('Distribution of Survived', size= 16);
plt.xlabel('Survived', size= 12);
```

Pclass wrt. Survived

```python
sns.countplot(data= titanic, x= 'Pclass', hue= 'Survived');
plt.title('Pclass wrt. Survived', size= 16);
plt.xlabel('Survived', size= 12);
```

SibSp wrt. Survived

```python
sns.set_style('darkgrid')
```

```python
sns.countplot(data= titanic, x= 'SibSp', hue= 'Survived');
plt.title('SibSp wrt. Survived', size= 16);
plt.xlabel('Survived', size= 12);
```

Parch wrt. Survived

```python
sns.countplot(data= titanic, x= 'Parch', hue= 'Survived');
plt.title('Parch wrt. Survived', size= 16);
plt.xlabel('Survived', size= 12);
```

Embarked wrt. Survived

```python
sns.countplot(data= titanic, x= 'Embarked', hue= 'Survived');
plt.title('Embarked wrt. Survived', size= 16);
plt.xlabel('Embarked', size= 12);
```

Distribution of Age

```python
sns.kdeplot(data= titanic, x= 'Age', color= 'r');
plt.title('Distribution of Age', size= 16);
plt.xlabel('Age', size= 12);
```
