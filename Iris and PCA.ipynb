```python
from sklearn.datasets import load_iris

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

### Load Data

```python
iris = load_iris(as_frame= True)
```

```python
df = iris.frame
df.head()
```

### Visualizations

Sepal length vs Sepal width

```python
sns.scatterplot(data= df, x= 'sepal length (cm)', y= 'sepal width (cm)', hue= 'target');
plt.title('Sepal length vs Sepal width', size= 14);
```

Petal length vs Petal width

```python
sns.scatterplot(data= df, x= 'petal length (cm)', y= 'petal width (cm)', hue= 'target');
plt.title('Petal length vs Petal width', size= 14);
```

The classes visually separate well in the second plot. Class 1 and class 2 does not have a clear visual separation.

### PCA

#### Find different basis instead of the sepal length and sepal width. Project the data (2-dimensional) to the new space spanned by the principal components

PCA instance

```python
pca = PCA(n_components= 2)
```

Scale the two variables (as to not interfere with PCA - high variance in 1 dimension can skew the result)

```python
sepal_scaled = StandardScaler().fit_transform(df.iloc[:, [0,1]])
```

Apply PCA transformation

```python
rotated = pca.fit_transform(sepal_scaled)
```

#### Visualize data with new bases

```python
sns.scatterplot(x= rotated[:,0], y= rotated[:,1], hue= df.target);
plt.title('Principal comp 1 vs Principal comp 2', size= 14);
```

The rotation did not help to get a better seperation (visual).
